```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.spatial import Voronoi, voronoi_plot_2d
```

```python
n = 1000
dim = 2
x = np.random.random((n, dim))
centers = x[np.random.choice(x.shape[0], 10)]
print(x.shape)
print(centers.shape)
```

```python
vor = Voronoi(centers)
```

```python
fig = voronoi_plot_2d(vor)
plt.scatter(x[:,0], x[:, 1], marker="x", color="lightgrey")
plt.show()
```
