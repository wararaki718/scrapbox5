```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(42)
```

```python
def model(x: float, beta0: float=0.3, beta1: float=0.5) -> float:
    y = beta0 * x + beta1
    return y
```

```python
noise = np.random.normal(0, 3, 50)
x = np.arange(0, len(noise))
```

```python
w1 = 0.3
w2 = 0.5
y = np.array([model(xi, w1, w2)+nz for xi, nz in zip(x, noise)])
plt.scatter(y, range(len(y)))
plt.show()
```

```python
# oridinary least square
a = np.dot(x, y) / (x**2).sum()
b = y.mean() - a*x.mean()
print(a)
print(b)
```

```python
w1 = 0.3
w2 = 0.5
y_ = np.array([model(xi, a, b) for xi in range(len(x))])
plt.scatter(range(len(y)), y)
plt.plot(range(len(y)), y_)
plt.show()
```
